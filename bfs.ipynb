```python
class Graph_search():
    def __init__(self, graph, type, index):
        self.type = type
        self.index = index
        self.visited = set()
        self.graph = graph # To avoid error
        if type == 'adj_matrix':
            self.graph = self.convert_adj_matrix()
        else:
            self.graph = graph

    
    def convert_adj_matrix(self):
        new_dict = dict()
        # First create linked list with only keys
        for i in range(len(self.graph)):
            temp_list = []
            if self.index == 1:
                new_dict[str(i + 1)] = temp_list
            else:
                new_dict[str(i)] = temp_list
        # Second append to the linked list if there was an element
        for index, rows in enumerate(self.graph):
            for element in rows:
                if element != 0:
                    keys = index + 1
                    new_dict[str(keys)].append(str(element))
        return new_dict

    def BFS(self, s):
        visited = set()
        queue   = set()      #list in python is basically queue
        visited.add(s) #means make it black
        queue.add(s)
        
        while queue:      #as long as the queue is not empty....
            u = queue.pop()  #pop the front guy.....basically index 0
            
            print(u, "-->", end = "")
            
            for neighbor in self.graph[u]: #for everyone who connects to u, 
                if neighbor not in visited:
                    visited.add(neighbor)              #add them to the visited
                    queue.add(neighbor)                #add them to the queue
    
    def DFS(self, s):
        if s not in self.visited:
            print(s, "--->", end="")  #our answer
            self.visited.add(s)
            for neighbor in self.graph[s]:
                self.DFS(neighbor)

```

```python
# Generate adjacency Matrix
# Modify code from https://www.delftstack.com/howto/python/python-adjacency-matrix/#:~:text=An%20adjacency%20matrix%20consists%20of,node%20j%20are%20not%20connected.
def generate_adjatency_matrix(edges, row_num=7, column_num=7, undirectional=False):
    #row_num = 7
    #column_num = 7
    adjatency_matrix = []
    for i in range(row_num):
        row = []
        for j in range(column_num):
            row.append(0)
        adjatency_matrix.append(row)
        #print(f'test {adjatency_matrix}')
    #edges = [(1, 2), (2, 7), (2, 4), (3, 4), (4, 6), (5, 4), (5, 3), (7, 5)]
    for edge in edges:
        row = edge[0]
        col = edge[1]
        adjatency_matrix[row - 1][col - 1] = col
        if undirectional != False:
            adjatency_matrix[col - 1][row - 1] = row # Uncommented this for undirectional graph

    #print(adjatency_matrix)
    return adjatency_matrix

    '''
    Matrix look likes this (undirectional graph)
    1   2  3  4  5  6  7
    1  [0, 2, 0, 0, 0, 0, 0]
    2  [1, 0, 0, 4, 0, 0, 7]
    3  [0, 0, 0, 4, 5, 0, 0]
    4  [0, 2, 3, 0, 5, 6, 0]
    5  [0, 0, 3, 4, 0, 0, 7]
    6  [0, 0, 0, 4, 0, 0, 0]
    7  [0, 2, 0, 0, 5, 0, 0]
    '''

adjatency_matrix =generate_adjatency_matrix([(1, 2), (2, 7), (2, 4), (3, 4), (4, 6), (5, 4), (5, 3), (7, 5)], 7, 7)
```
